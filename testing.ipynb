```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

x,y = np.meshgrid(np.arange(0,100,1),np.arange(0,100,1))
pos = np.dstack((x,y))
means = np.array([50,50])
cov = np.array([[250,0],[0,250]])
rv = multivariate_normal(means,cov)
prob = rv.pdf(pos)

plt.imshow(prob)
print(prob.shape)
print(np.sum(prob))

# x,y = np.random.default_rng().multivariate_normal(means,cov,10000).T
# plt.scatter(x,y)
# hist = np.histogram2d(x,y,100)
# plt.imshow(hist[0])
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(100)
y = np.arange(100)
arr = np.zeros((x.size,y.size),float)

fig,[ax1,ax2] = plt.subplots(1,2)
ax1.imshow(arr)


for i in range(8):
    cx = np.random.choice(x)
    cy = np.random.choice(y)
    r = np.random.choice([4,6,8,10])
    arr[(x[np.newaxis,:]-cx)**2+(y[:,np.newaxis]-cy)**2 < r**2] = 1


ax2.imshow(arr)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter
arr = np.load('mcarlo_sim.npy')
w_x = np.arange(0,1,0.02)
_ = plt.plot(w_x,arr.T)
w_std = np.std(arr,axis=0)
plt.xlabel('Single Scattering Albedo')
plt.ylabel('Reflectance')
plt.savefig('Monte-Carlo_Result.png')
fig,ax = plt.subplots()
ax.yaxis.set_major_formatter(FormatStrFormatter('%.0E'))
ax.plot(w_x,w_std)
ax.set_xlabel('Single Scattering Albedo')
ax.set_ylabel('Monte-Carlo 1\u03c3 Deviation')
plt.savefig('Monte-Carlo_Deviation.png')
```
